```python
import os
ROOT = '/content/drive/MyDrive/stoch_depth'
for folder in [
    'code',
    'fig3/stochastic/checkpoints', 'fig3/stochastic/logs',
    'fig3/baseline/checkpoints',   'fig3/baseline/logs',
    'fig7/stochastic/checkpoints', 'fig7/stochastic/logs',
    'fig7/baseline/checkpoints',   'fig7/baseline/logs',
    'fig8_left/linear/checkpoints', 'fig8_left/linear/logs',
    'fig8_left/uniform/checkpoints','fig8_left/uniform/logs',
    'fig8_right/checkpoints',       'fig8_right/logs',
    'figures',
]:
    os.makedirs(f'{ROOT}/{folder}', exist_ok=True)
print('Folders created.')
```

```python
from google.colab import files
uploaded = files.upload()  # select model.py, data.py, train.py, evaluate.py
```

```python
import shutil, os
ROOT = '/content/drive/MyDrive/stoch_depth'
for fname in ['model.py', 'data.py', 'train.py', 'evaluate.py']:
    src = f'/content/{fname}'
    if os.path.exists(src):
        shutil.copy(src, f'{ROOT}/code/{fname}')
        print(f'Saved {fname} to Drive')
    else:
        print(f'WARNING: {fname} not found')
```

```python
ROOT = '/content/drive/MyDrive/stoch_depth'
for fname in ['model.py', 'data.py', 'train.py', 'evaluate.py']:
    status = 'OK' if os.path.exists(f'{ROOT}/code/{fname}') else 'MISSING'
    print(f'{status}  {fname}')
print('Setup complete.')
```
